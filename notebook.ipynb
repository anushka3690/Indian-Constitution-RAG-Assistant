```python
import fitz
doc = fitz.open("Constitution of India.pdf")
```

```python
with open("output.txt", "w", encoding="utf-8") as f:
    for page in doc:
        text = page.get_text()
        f.write(text)
        f.write("\n")

print("Text saved to output.txt")
```
